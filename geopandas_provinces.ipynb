# Geopandas Visualización de provincias del Perú

## 1.- Carga de librerias

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

## 2.- Carga de Archivo Geojson que contiene polígonos de las provincias del Perú

Fuente Oficial de Geojson usado en este Notebook: [Peru_Geojson](https://github.com/juaneladio/peru-geojson)

```python
url_geojson = "geojson/peru_provincial_simple.geojson"
```

```python
provinces_geojson = gpd.read_file(url_geojson)
provinces_geojson.head()
```

Como se puede observar el archivo geojson viene con los poligonos de las provincias pero también con la columna FIRST_NOMB que indica a la region que pertenece.

## 2.- Visualizacion


### 2.1 Visualización Simple

```python
ax = provinces_geojson.plot(figsize=(20,20))
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

### 2.2 Visualización a color

```python
ax = provinces_geojson.plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

## 2.3 Mapa Personalizado


#### Mostrar una Region con sus provincias. Ejm. **Cusco**

```python
ax = provinces_geojson[provinces_geojson.FIRST_NOMB=='CUSCO'].plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```

#### Mostrar una provincia. Ejm. **Calca**

```python
ax = provinces_geojson[provinces_geojson.NOMBPROV=='CALCA'].plot(figsize=(10,10),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```
