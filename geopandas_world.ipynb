# Geopandas Visualización del Mapa Mundial

## 1.- Carga de librerias

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

## 2.- Carga de Archivo Geojson que contiene polígonos de los paises del mundo

Fuente Oficial de Geojson usado en este Notebook: [Geo-Countries](https://github.com/datasets/geo-countries)

```python
url_geojson = "geojson/countries.geojson"
```

```python
world_geojson = gpd.read_file(url_geojson)
world_geojson.head()
```

## 2.- Visualizacion


### 2.1 Visualización Simple

```python
ax = world_geojson.plot(figsize=(20,20))
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

### 2.2 Visualización a color

```python
ax = world_geojson.plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

## 2.3 Mapa Personalizado


#### Excuir un pais. Ejm. **Antarctica**

```python
ax = world_geojson[world_geojson.ADMIN!='Antarctica'].plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```

#### Mostrar un único pais. Ejm. **Peru**

```python
ax = world_geojson[world_geojson.ADMIN=='Peru'].plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```
