# Geopandas Visualización de los distritos del Perú

## 1.- Carga de librerias

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

## 2.- Carga de Archivo Geojson que contiene polígonos de los distritos del Perú

Fuente Oficial de Geojson usado en este Notebook: [Peru_Geojson](https://github.com/juaneladio/peru-geojson)

```python
url_geojson = "geojson/peru_distrital_simple.geojson"
```

```python
districts_geojson = gpd.read_file(url_geojson)
districts_geojson.head()
```

## 2.- Visualizacion


### 2.1 Visualización Simple

```python
ax = districts_geojson.plot(figsize=(20,20))
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

### 2.2 Visualización a color

```python
ax = districts_geojson.plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

## 2.3 Mapa Personalizado


#### Mostrar una Region con sus distritos. Ejm. **Cusco**

```python
ax = districts_geojson[districts_geojson.NOMBDEP=='CUSCO'].plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```

#### Mostrar una provincia. Ejm. **Calca**

```python
ax = districts_geojson[districts_geojson.NOMBPROV=='CALCA'].plot(figsize=(10,10),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```

#### Mostrar una distrito. Ejm. **Pisac**

```python
ax = districts_geojson[districts_geojson.NOMBDIST=='YANATILE'].plot(figsize=(10,10),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```

Podemos observar que incluso se puede graficar el distrito, pero cabe mencionar que el polígono no dibuja los detalles finos que tiene los límtes fronterizos del distrito.
