# Geopandas Visualización de regiones del Perú

## 1.- Carga de librerias

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

## 2.- Carga de Archivo Geojson que contiene polígonos de las regiones del Perú

Fuente Oficial de Geojson usado en este Notebook: [Peru_Geojson](https://github.com/juaneladio/peru-geojson)

```python
url_geojson = "geojson/peru_departamental_simple.geojson"
```

```python
region_geojson = gpd.read_file(url_geojson)
region_geojson.head()
```

## 2.- Visualizacion


### 2.1 Visualización Simple

```python
ax = region_geojson.plot(figsize=(20,20))
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

### 2.2 Visualización a color

```python
ax = region_geojson.plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

## 2.3 Mapa Personalizado


#### Mostrar una Region. Ejm. **Cusco**

```python
ax = region_geojson[region_geojson.NOMBDEP=='CUSCO'].plot(figsize=(20,20),edgecolor=u'gray', cmap='Pastel1')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
ax.axis('scaled')
plt.show()
```
